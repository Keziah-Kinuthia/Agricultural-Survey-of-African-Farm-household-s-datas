```python
farmwork_cols = [x for x in df.columns.to_list() if x.startswith("farmwork")]
nfarmwork_cols = [x for x in df.columns.to_list() if x.startswith("nfarmwork")]
```

```python
# Check if any household member works on farm
df['works_on_farm'] = df[farmwork_cols].any(axis=1)

# Check if any household member works on non-farm activities
df['works_on_nonfarm'] = df[nfarmwork_cols].any(axis=1)

# Create a new column to check if a household member is engaged in both activities
df['engaged_in_both'] = df['works_on_farm'] & df['works_on_nonfarm']

# Calculate the number of households with at least one member engaged in both activities
num_households_with_both = df['engaged_in_both'].sum()

# Calculate the total number of households
total_households = len(df)

# Calculate the probability
probability_both = num_households_with_both / total_households

print(f"Probability that a household has at least one member engaged in both farm and non-farm activities: {probability_both}")
```

```python
# Sample size
n = 100

# Expected number of households meeting the criterion
expected_households = n * probability_both

print(f"Expected number of households with at least one member engaged in both farm and non-farm activities in a sample of 100 households: {expected_households}")
```

```python
marital_status_cols = [x for x in df.columns.to_list() if x.startswith("married")]
```

```python
print(len(marital_status_cols))
```

```python
# Combine all marital status columns into a single series
marital_status_combined = pd.Series(dtype='int')
marital_status_combined = pd.concat([df[col] for col in marital_status_cols], ignore_index=True)

# Map the numerical values to their corresponding categories
marital_status_map = {
    1: 'Married',
    2: 'Never Married',
    3: 'Previously Married',
    4: 'Not Applicable'
}
marital_status_combined = marital_status_combined.map(marital_status_map)

# Count the frequency of each marital status category
marital_status_counts = marital_status_combined.value_counts()

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x=marital_status_counts.index, y=marital_status_counts.values, palette='viridis')

# Add titles and labels
plt.title('Frequency Distribution of Marital Status Categories')
plt.xlabel('Marital Status')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
